```python
# Complete the code below :

# D√©finition de la figure
# Le param√®tre figsize permet de d√©finir la largeur et la hauteur de la figure
fig, ax = plt.subplots(figsize = (20,10))

# D√©finition du premier graphique
ax1 = plt.subplot(234) 
ax1.scatter(x = df_weather["DATE"], y = df_weather["MAX_TEMPERATURE_C"], c= 'red')
ax1.set_yticks(range(-5,45, 5))
plt.ylabel("Temperature MAX")
ax1.set_title("Temperature MAX")

# D√©finition du second graphique
ax2 = plt.subplot(235) 
ax2.scatter(x = df_weather["DATE"], y = df_weather["MIN_TEMPERATURE_C"], c= 'blue')
ax2.set_yticks(range(-5,45, 5))
plt.ylabel("Temperature MIN")
ax2.set_title("Temperature MIN")

plt.show()
```
