## Mission 4

Create a function using Numpy that solves [second degree equations of this kind](https://www.youtube.com/watch?v=hwq_zpsGixc): $ax^2+bx+c$

```
import numpy as np

def solve_quadratic(a, b, c):
    """
    Solves a quadratic equation of the form ax^2 + bx + c = 0
    """
    # Calculate the discriminant
    discriminant = b**2 - 4*a*c
    
    # Find the square root of the discriminant using numpy
    sqrt_discriminant = np.sqrt(discriminant)
    
    # Check if the discriminant is positive, zero or negative 
    if discriminant > 0:
        # Real and distinct roots
        root1 = (-b + sqrt_discriminant) / (2*a)
        root2 = (-b - sqrt_discriminant) / (2*a)
        return root1, root2
    elif discriminant == 0:
        # Real and equal roots
        root = -b / (2*a)
        return root
    else:
        # Complex roots
        real = -b / (2*a)
        imag = sqrt_discriminant / (2*a)
        root1 = complex(real, imag)
        root2 = complex(real, -imag)
        return root1, root2

# Example usage
a = 2
b = -3
c = 1
result = solve_quadratic(a, b, c)
print(result)
```

Re-code the Gaussian curve using NumPy functions instead of Python functions.

```
# Code with NumPy

import numpy as np
import matplotlib.pyplot as plt

# Domain of x values
domain = np.linspace(-100, 100, 200)

# Mean and standard deviation
mu = 0
sigma = 20

# Gaussian function using NumPy
y = 1/(np.sqrt(2*np.pi*np.power(sigma,2))) * np.exp(-np.power((domain-mu),2)/(2*np.power(sigma,2)))

# Plot the function
plt.plot(domain, y)
plt.show()
```
